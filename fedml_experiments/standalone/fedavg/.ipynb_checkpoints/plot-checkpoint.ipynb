```python
infile = r"./LOG/cifar10/fedavg-dir0.3-mdlresnet18-cm50-total_clnt10-neighbor1-seed2022.log"

important = []
keep_phrases = ["global_test_acc",
              "global_test_loss"]

with open(infile) as f:
    f = f.readlines()

for line in f:
    for phrase in keep_phrases:
        if phrase in line:
            line = line.replace('\n', '')
            important.append(line)
            break
```

```python
import json
def get_acc_and_loss(important):
    global_test_acc=[]
    global_test_loss=[]
    for data in important:
        data = data.replace("'", '"')
        j = json.loads(data)
        global_test_acc.append(j['global_test_acc'])
        global_test_loss.append(j['global_test_loss'])
    return global_test_acc,global_test_loss
```

```python
import matplotlib
import matplotlib.pyplot as plt
# Plot Loss curve
global_test_acc,global_test_loss=get_acc_and_loss(important)
plt.figure()
plt.title('Training Loss vs Communication rounds')
plt.plot(range(len(global_test_loss)), global_test_loss, color='r',label='FedAvg')
plt.ylabel('Training loss')
plt.xlabel('Communication Rounds')
plt.legend()
plt.show()
```

```python
# Plot Average Accuracy vs Communication rounds
plt.figure()
plt.title('Average Accuracy vs Communication rounds')
plt.plot(range(len(global_test_acc)), global_test_acc, color='r',label='FedAvg')
plt.ylabel('Average Accuracy')
plt.xlabel('Communication Rounds')
plt.legend()
plt.show()
```
